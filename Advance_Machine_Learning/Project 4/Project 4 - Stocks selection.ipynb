```python
import pandas as pd

file_path = '/Users/sieng/Documents/Study/MS.Business Analytics/FALL 2024/AML/Final Project/SP500_data.csv'

sp500_df = pd.read_csv(file_path)
sp500_df
```

```python
## remove NA
count_na = sp500_df.isna().sum().sum()
print(count_na)
count_na_row = sp500_df.isna().sum()
print(count_na_row)
sp500_cleaned = sp500_df.dropna()
sp500_cleaned
```

```python
sp500_buy = sp500_cleaned.drop(sp500_df.columns[[0, 1, 2, 4]], axis=1)
sp500_sell = sp500_cleaned.drop(sp500_df.columns[[0, 1, 2, 3]], axis=1)
sp500_return = sp500_cleaned.drop(sp500_df.columns[[0, 1, 3, 4]], axis=1)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separate features and target
X_buy = sp500_buy.iloc[:, 1:] # All columns except first column
y_buy = sp500_buy.iloc[:, 0] # First column as target

# Split the data into training and testing sets
X_buy_train, X_buy_test, y_buy_train, y_buy_test = train_test_split(X_buy, y_buy, test_size=0.3, random_state=22)

# Normalize features using StandardScaler (normal curve normalization)
scaler = StandardScaler()
X_buy_train_scaled = scaler.fit_transform(X_buy_train)
X_buy_test_scaled = scaler.transform(X_buy_test)

# Count occurrences of each class in y_buy_test using numpy
classes_buy, counts_buy = np.unique(y_buy_train, return_counts=True)
class_counts_buy = dict(zip(classes_buy, counts_buy))
print("Class distribution in y_buy_test (using numpy):", class_counts_buy)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow import keras
from tensorflow.keras import layers, regularizers
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt


# Create early stopping callback
model_buy_callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath = "best_model_buy.keras",
        save_best_only=True,
        monitor="val_loss"  # Tracks the validation loss
    ),
    keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=5,  # Stop if no improvement after 5 epochs
        restore_best_weights=True
    )
]

# Build the Neural Network model
model_buy = keras.Sequential([
    # Input layer
    layers.Input(shape=(X_buy_train.shape[1],)),
    
    # Hidden layers with dropout for regularization
    layers.Dense(32, activation="relu", kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.2),
    
    # Output layer
    layers.Dense(1, activation="sigmoid")
])

# Compile the model
model_buy.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy',
             keras.metrics.Precision(), 
             keras.metrics.Recall()]
)

# Train the model with early stopping callback
history_buy = model_buy.fit(
    X_buy_train_scaled, y_buy_train,
    validation_split=0.2,  # Use 20% of training data for validation
    epochs=100,
    batch_size=32,
    callbacks=model_buy_callbacks,  # Add early stopping callback
    verbose=1
)

# Print the number of epochs actually used
print(f"\nTraining stopped after {len(history_buy.history['loss'])} epochs")

# Load the best model saved during training
best_model_buy = keras.models.load_model("best_model_buy.keras")

# Get raw probabilities from the model
y_buy_probabilities = best_model_buy.predict(X_buy_test_scaled)

# Calculate the average raw probability
average_buy_probability = np.mean(y_buy_probabilities)
print(f"Average Raw Probability of the Predictions: {average_buy_probability:.4f}")

# Evaluate the model
y_buy_pred = (best_model_buy.predict(X_buy_test_scaled) > average_buy_probability).astype(int)

# Optional: Plot training history
plt.figure(figsize=(12,4))

# Plot training & validation accuracy values
plt.subplot(1,2,1)
plt.plot(history_buy.history['accuracy'])
plt.plot(history_buy.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1,2,2)
plt.plot(history_buy.history['loss'])
plt.plot(history_buy.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()
plt.show()

# Generate and display confusion matrix
cm_buy = confusion_matrix(y_buy_test, y_buy_pred)
disp_buy = ConfusionMatrixDisplay(confusion_matrix=cm_buy, display_labels=["Class 0", "Class 1"])
disp_buy.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()

# Predict probabilities for ROC curve
y_buy_pred_proba = best_model_buy.predict(X_buy_test_scaled).flatten()

# Calculate ROC curve and AUC
fpr_buy, tpr_buy, thresholds_buy = roc_curve(y_buy_test, y_buy_pred_proba)
roc_auc_buy = auc(fpr_buy, tpr_buy)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_buy, tpr_buy, color='darkorange', lw=2, 
         label=f'ROC curve (AUC = {roc_auc_buy:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Print additional ROC metrics
print(f"\nROC AUC Score: {roc_auc_buy:.4f}")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow import keras
from tensorflow.keras import layers, regularizers
import matplotlib.pyplot as plt

def create_improved_model(input_shape):
    model_buy3 = keras.Sequential([
        # Input layer
        layers.Input(shape=(input_shape,)),
        
        # Batch normalization and dropout for improved generalization
        layers.BatchNormalization(),
        
        # Hidden layers with more sophisticated regularization
        layers.Dense(64, activation="relu", 
                     kernel_regularizer=regularizers.l2(0.0005),
                     kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(128, activation="relu", 
                     kernel_regularizer=regularizers.l2(0.0005),
                     kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.Dropout(0.4),
        
        layers.Dense(256, activation="relu", 
                     kernel_regularizer=regularizers.l2(0.0005),
                     kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.Dropout(0.4),
        
        layers.Dense(128, activation="relu", 
                     kernel_regularizer=regularizers.l2(0.0005),
                     kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        # Output layer
        layers.Dense(1, activation="sigmoid")
    ])
    
    return model_buy3

# Learning rate scheduler
def lr_schedule(epoch):
    initial_lr = 0.001
    drop = 0.5
    epochs_drop = 10.0
    lr = initial_lr * (drop ** np.floor((1+epoch)/epochs_drop))
    return lr

# Callbacks with more sophisticated early stopping and learning rate scheduling
model_buy_callbacks3 = [
    keras.callbacks.ModelCheckpoint(
        filepath="best_model_sell3.keras", 
        save_best_only=True, 
        monitor="val_loss"
    ),
    keras.callbacks.EarlyStopping(
        monitor="val_loss", 
        patience=5,  # Increased patience
        restore_best_weights=True,
        min_delta=0.001  # Minimum change to qualify as an improvement
    ),
    keras.callbacks.LearningRateScheduler(lr_schedule)
]

# Create and compile the model
model_buy3 = create_improved_model(X_buy_train_scaled.shape[1])
model_buy3.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy', 
    metrics=['accuracy', 
             keras.metrics.Precision(), 
             keras.metrics.Recall()]
)

# Train the improved model
history_buy3 = model_buy3.fit(
    X_buy_train_scaled, y_buy_train, 
    validation_split=0.2,
    epochs=50,  # Increased max epochs
    batch_size=32, 
    callbacks=model_buy_callbacks3,
    verbose=1
)

# Load the best model
best_model_buy3 = keras.models.load_model("best_model_buy3.keras")

# Print training details
print(f"\nTraining stopped after {len(history_buy3.history['loss'])} epochs")

# Get raw probabilities from the model
y_buy_probabilities3 = best_model_buy3.predict(X_buy_test_scaled)

# Calculate the average raw probability
average_buy_probability3 = np.mean(y_buy_probabilities3)
print(f"Average Raw Probability of the Predictions: {average_buy_probability3:.4f}")

# Evaluate the model
y_buy_pred3 = (best_model_buy3.predict(X_buy_test_scaled) > average_buy_probability3).astype(int)

# Plotting training history
plt.figure(figsize=(15,5))

# Accuracy plot
plt.subplot(1,3,1)
plt.plot(history_buy3.history['accuracy'], label='Train Accuracy')
plt.plot(history_buy3.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

# Loss plot
plt.subplot(1,3,2)
plt.plot(history_buy3.history['loss'], label='Train Loss')
plt.plot(history_buy3.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

# Learning Rate plot
plt.subplot(1,3,3)
lr_history = [lr_schedule(epoch) for epoch in range(len(history_buy3.history['loss']))]
plt.plot(lr_history)
plt.title('Learning Rate')
plt.ylabel('Learning Rate')
plt.xlabel('Epoch')
plt.tight_layout()
plt.show()

# Generate and display confusion matrix
cm_buy3 = confusion_matrix(y_buy_test, y_buy_pred3)
disp_buy3 = ConfusionMatrixDisplay(confusion_matrix=cm_buy3, display_labels=["Class 0", "Class 1"])
disp_buy3.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()

# Predict probabilities for ROC curve
y_buy_pred_proba3 = best_model_buy3.predict(X_buy_test_scaled).flatten()

# Calculate ROC curve and AUC
fpr_buy3, tpr_buy3, thresholds_buy3 = roc_curve(y_buy_test, y_buy_pred_proba3)
roc_auc_buy3 = auc(fpr_buy3, tpr_buy3)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_buy3, tpr_buy3, color='darkorange', lw=2, 
         label=f'ROC curve (AUC = {roc_auc_buy3:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Print additional ROC metrics
print(f"\nROC AUC Score: {roc_auc_buy3:.4f}")
```
